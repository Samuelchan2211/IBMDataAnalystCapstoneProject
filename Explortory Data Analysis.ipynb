## Hands on Lab


##### Step 1: Install and Import Required Libraries


```python
# Install the necessary libraries
!pip install pandas
!pip install matplotlib
!pip install seaborn

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2: Load the Dataset


```python
# Load the dataset from the given URL
file_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_url)

# Display the first few rows to understand the structure of the dataset
df.head()
```

<h3>Step 3: Analyze and Visualize Compensation Distribution</h3>


**Task**: Plot the distribution and histogram for `ConvertedCompYearly` to examine the spread of yearly compensation among respondents.


```python
## Write your code here
# Drop NaN values to avoid plotting issues
comp_data = df['ConvertedCompYearly'].dropna()

# Set up the figure
plt.figure(figsize=(12, 6))

# Histogram with KDE (Kernel Density Estimate)
sns.histplot(comp_data, bins=50, kde=True, color='skyblue')

# Add titles and labels
plt.title('Distribution of Yearly Compensation (ConvertedCompYearly)', fontsize=16)
plt.xlabel('Yearly Compensation (USD)', fontsize=12)
plt.ylabel('Number of Respondents', fontsize=12)

# Show the plot
plt.show()
```

<h3>Step 4: Calculate Median Compensation for Full-Time Employees</h3>


**Task**: Filter the data to calculate the median compensation for respondents whose employment status is "Employed, full-time."


```python
## Write your code here
# Filter the data for full-time employed respondents
full_time_df = df[df['Employment'] == 'Employed, full-time']

# Calculate the median of ConvertedCompYearly
median_comp = full_time_df['ConvertedCompYearly'].median()

# Display the result
print("Median Yearly Compensation for Full-Time Employees: $", round(median_comp, 2))
```

<h3>Step 5: Analyzing Compensation Range and Distribution by Country</h3>


Explore the range of compensation in the ConvertedCompYearly column by analyzing differences across countries. Use box plots to compare the compensation distributions for each country to identify variations and anomalies within each region, providing insights into global compensation trends.



```python
## Write your code here
# Drop rows with missing compensation or country
df_filtered = df[['Country', 'ConvertedCompYearly']].dropna()

# Optional: Filter to countries with at least 50 responses for clarity
country_counts = df_filtered['Country'].value_counts()
valid_countries = country_counts[country_counts >= 50].index
df_filtered = df_filtered[df_filtered['Country'].isin(valid_countries)]
plt.figure(figsize=(16, 8))

# Create the box plot
sns.boxplot(x='Country', y='ConvertedCompYearly', data=df_filtered)
plt.xticks(rotation=45, ha='right')
plt.title('Yearly Compensation Distribution by Country')
plt.xlabel('Country')
plt.ylabel('Yearly Compensation (USD)')
plt.tight_layout()
plt.show()
import numpy as np

# Add a log-transformed column
df_filtered['LogComp'] = np.log1p(df_filtered['ConvertedCompYearly'])

plt.figure(figsize=(16, 8))
sns.boxplot(x='Country', y='LogComp', data=df_filtered)
plt.xticks(rotation=45, ha='right')
plt.title('Log-Transformed Compensation Distribution by Country')
plt.xlabel('Country')
plt.ylabel('Log(1 + Yearly Compensation)')
plt.tight_layout()
plt.show()
```

<h3>Step 6: Removing Outliers from the Dataset</h3>


**Task**: Create a new DataFrame by removing outliers from the `ConvertedCompYearly` column to get a refined dataset for correlation analysis.


```python
## Write your code here
# Step 1: Drop missing values in the target column
comp_data = df['ConvertedCompYearly'].dropna()

# Step 2: Calculate IQR
Q1 = comp_data.quantile(0.25)
Q3 = comp_data.quantile(0.75)
IQR = Q3 - Q1

# Step 3: Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Step 4: Filter the original DataFrame
df_refined = df[(df['ConvertedCompYearly'] >= lower_bound) & (df['ConvertedCompYearly'] <= upper_bound)]

# Step 5: Preview the refined dataset
print("Original shape:", df.shape)
print("Refined shape:", df_refined.shape)
df_refined.head()
```

<h3>Step 7: Finding Correlations Between Key Variables</h3>


**Task**: Calculate correlations between `ConvertedCompYearly`, `WorkExp`, and `JobSatPoints_1`. Visualize these correlations with a heatmap.


```python
## Write your code here
# Select the relevant columns
corr_df = df[['ConvertedCompYearly', 'WorkExp', 'JobSatPoints_1']].dropna()
# Compute the correlation matrix
correlations = corr_df.corr()
print(correlations)
import seaborn as sns
import matplotlib.pyplot as plt

# Set up the plot
plt.figure(figsize=(8, 6))
sns.heatmap(correlations, annot=True, cmap='coolwarm', fmt=".2f")

# Add titles and labels
plt.title('Correlation Heatmap: Compensation, Experience & Job Satisfaction')
plt.show()
```

<h3>Step 8: Scatter Plot for Correlations</h3>


**Task**: Create scatter plots to examine specific correlations between `ConvertedCompYearly` and `WorkExp`, as well as between `ConvertedCompYearly` and `JobSatPoints_1`.


```python
## Write your code here
import matplotlib.pyplot as plt
import seaborn as sns

# Drop missing values for both columns
df_exp = df[['ConvertedCompYearly', 'WorkExp']].dropna()

# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_exp, x='WorkExp', y='ConvertedCompYearly', alpha=0.6)
plt.title('Compensation vs Work Experience')
plt.xlabel('Years of Work Experience')
plt.ylabel('Yearly Compensation (USD)')
plt.grid(True)
plt.show()
# Drop missing values for both columns
df_sat = df[['ConvertedCompYearly', 'JobSatPoints_1']].dropna()

# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_sat, x='JobSatPoints_1', y='ConvertedCompYearly', alpha=0.6)
plt.title('Compensation vs Job Satisfaction')
plt.xlabel('Job Satisfaction Score')
plt.ylabel('Yearly Compensation (USD)')
plt.grid(True)
plt.show()
import numpy as np

df_exp['LogComp'] = np.log1p(df_exp['ConvertedCompYearly'])
sns.scatterplot(data=df_exp, x='WorkExp', y='LogComp')
plt.title('Log-Transformed Compensation vs Work Experience')
plt.xlabel('Years of Work Experience')
plt.ylabel('Log(1 + Compensation)')
plt.show()
```
